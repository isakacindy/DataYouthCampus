##12세 미만

```python
child = pd.read_csv('/content/drive/MyDrive/데이터 청년 캠퍼스/클러스터링 데이터/2020_2021_2022_만12세이하_전처리.csv', encoding = 'cp949')
child
```

```python
child = child.drop(['2020년12월_거주자_연령구간인구수'], axis = 1)
```

```python
child = child.drop(['2021년12월_거주자_연령구간인구수'], axis = 1)
new_names = {'2022년12월_거주자_연령구간인구수': '2022년_만12세이하_인구수'}
child.rename(columns=new_names, inplace=True) # 22년 인구수 사용
child
```

##데이터 병합

```python
merged_df = pd.merge(old, child, on=['자치구','행정동']) # 자치구 행정동으로 병합
merged_df
```

```python
merged_df = pd.merge(merged_df, handicap, on=['자치구','행정동']) # 자치구 행정동으로 병합
merged_df
```

```python
merged_df = pd.merge(merged_df, dust, on=['자치구','행정동']) #  자치구 행정동으로 병합
merged_df
```

```python
merged_df = pd.merge(merged_df, people_year, on=['자치구','행정동']) #  자치구 행정동으로 병합
```

```python
merged_df = pd.merge(merged_df, hot_str, on=['자치구','행정동']) #  자치구 행정동으로 병합
```

```python
merged_df = pd.merge(merged_df, hot_fre, on=['자치구','행정동']) #  자치구 행정동으로 병합
```

```python
merged_df
```

```python
merged_df.rename(columns={'합계': '장애인'}, inplace=True) # 열 이름 변경 합계 -> 장애인

merged_df
```

```python
# "2022년_만65세이상_인구수", "2022년_만12세이하_인구수", "장애인" 값을 합쳐서 "교통약자" 컬럼 생성
merged_df['교통약자'] = merged_df['2022년_만65세이상_인구수'] + merged_df['2022년_만12세이하_인구수'] + merged_df['장애인']

# 결과 출력
merged_df
```

##점검

```python
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'NanumGothic'
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 변수들 간의 상관계수 계산
correlation_matrix = merged_df[['2022년_만65세이상_인구수', '2022년_만12세이하_인구수', '장애인','지체','시각',
                                '(초)미세먼지 나쁨 빈도', '연일일평균총승하차승객수', '폭염강도', '폭염빈도']].corr()

# 히트맵 생성
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('변수별 상관관계')
plt.xticks(rotation=30, ha='right', fontsize=10)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 변수들 간의 상관계수 계산
correlation_matrix = merged_df[['교통약자','(초)미세먼지 나쁨 빈도', '연일일평균총승하차승객수', '폭염빈도']].corr()

# 히트맵 생성
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('변수별 상관관계')
plt.xticks(rotation=30, ha='right', fontsize=10)
plt.show()
```
